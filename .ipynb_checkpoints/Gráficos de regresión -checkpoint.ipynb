```python
import seaborn as sns 
import matplotlib.pyplot as plt
% matplotlib inline 
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],
          scatter_kws={'s':100}) # podemos matizar con Hue y editar las formas con marker ,asi como aumentar de tamaño de puntos
```

```python
# Si deseo leer los datos en diferentes graficos debo usar como argumento columna y filas que basicamente serian 
# un FaceGrid
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',col='time',aspect=0.8,size=6)
```

# Estilos y colores en graficos (Estetica)

```python
sns.set_context('paper',font_scale=1) ## context : dict, None, or one of {paper, notebook, talk, poster
sns.set_style('whitegrid') # cambios de  style : dict, None, or one of {darkgrid, whitegrid, dark, white, ticks}
sns.countplot(x='sex',data=tips)
sns.despine() # la usamos para escoger que deseamos eliminar del grafico top=True, right=True, left=False, 
              # bottom=False, offset=None, trim=False
```

```python
# Grid Type Plot
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='seismic')
```
